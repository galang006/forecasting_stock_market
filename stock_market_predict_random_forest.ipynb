```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2200, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(bbri, model, predictors)
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = bbri.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    bbri[ratio_column] = bbri["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    bbri[trend_column] = bbri.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
bbri = bbri.dropna()
```

```python
model = RandomForestClassifier(
    n_estimators=1000, 
    min_samples_split=60, 
    min_samples_leaf=5,
    class_weight="balanced", 
    random_state=1
    )
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=1200, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(bbri, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
accuracy = accuracy_score(predictions["Target"], predictions["Predictions"])
precision = precision_score(predictions["Target"], predictions["Predictions"])
recall = recall_score(predictions["Target"], predictions["Predictions"])
f1 = f1_score(predictions["Target"], predictions["Predictions"])
```

```python
print(classification_report(predictions["Target"], predictions["Predictions"]))
```

```python
import matplotlib.pyplot as plt
vis_data = bbri.loc[predictions.index].copy()
vis_data["Prediction"] = predictions["Predictions"]
```

```python
plt.figure(figsize=(15, 6))

plt.plot(vis_data.index, vis_data["Close"], label="Harga Close", color='gray', alpha=0.5)

plt.scatter(
    vis_data.index[vis_data["Prediction"] == 1],
    vis_data["Close"][vis_data["Prediction"] == 1],
    color="green", label="Prediksi Naik", marker="^", s=60
)

plt.scatter(
    vis_data.index[vis_data["Prediction"] == 0],
    vis_data["Close"][vis_data["Prediction"] == 0],
    color="red", label="Prediksi Turun", marker="v", s=60
)

plt.title("Prediksi Model: Naik / Turun Harga Saham")
plt.xlabel("Tanggal")
plt.ylabel("Harga Penutupan (Close)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
